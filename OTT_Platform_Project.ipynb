#Number of Movies added on netflix in the past 20 years .


```python
# Convert "date_added" column to DateTime format
df['date_added'] = pd.to_datetime(df['date_added'])

# Filter the DataFrame for movies
movies_df = df[df['type'] == 'Movie']

# Calculate the number of movies added each year within the last 20 years
current_year = pd.Timestamp.now().year
twenty_years_ago = current_year - 20
movies_last_20_years = movies_df[movies_df['date_added'].dt.year >= twenty_years_ago]
movies_added_by_year = movies_last_20_years['release_year'].value_counts().sort_index()

# Create a scatter plot
plt.plot(movies_added_by_year.index, movies_added_by_year.values)
plt.xlabel('Release Year')
plt.ylabel('Number of Movies Added')
plt.title('Number of Movies Added to Netflix in the Last 20 Years')

# Display the plot
plt.show()
```

#hence it is infered that there was sharp rise in movies added on netflix after 2010 and till 2020 it has come down .

#What is the best time to launch a TV show


```python
#creating new data frame to filter out
df1= df[df['type']=='TV Show']

# Convert 'date_added' column to datetime type
df1['date_added'] = pd.to_datetime(df['date_added'])

# Extract the month from the 'date_added' column
df1['month_added'] = df['date_added'].dt.month

# Group by month and count the number of TV shows added
monthly_counts = df1.groupby('month_added')['type'].count()

# Plot the monthly counts
monthly_counts.plot(kind='line', marker='o', linestyle='-', color='red')

# Set labels for x-axis and y-axis
plt.xlabel('Month')
plt.ylabel('Number of TV Shows')

# Set x-tick labels
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
plt.xticks(range(1, 13), months)

# Show the plot
plt.show()

```

#The best time to launch a TV show is between june july or after Nov.

#Understanding what content is available in different countries

```python
df.head()
```

```python
df['listed_in'].value_counts()
```

```python




# Get the value counts of "listed_in"
listed_in_counts = df['listed_in'].value_counts()

# Create a new DataFrame with "listed_in" and corresponding count
count_df = pd.DataFrame({'listed_in': listed_in_counts.index, 'count': listed_in_counts.values})

# Get the top 10 countries based on the count of shows
top_10_countries = df['country'].value_counts().head(10).index

# Filter the DataFrame for the top 10 countries
filtered_df = df[df['country'].isin(top_10_countries)]

# Merge the count DataFrame with the filtered DataFrame based on "listed_in"
merged_df = pd.merge(filtered_df, count_df, on='listed_in')

# Create the scatter plot using Seaborn
sns.barplot(data=merged_df, x='country', y='count')

# Rotate x-axis labels for better visibility
plt.xticks(rotation=45, ha='right')

# Set labels for x-axis and y-axis
plt.xlabel('Country')
plt.ylabel('Count of Listed In')

# Set title for the plot
plt.title('top 10 Country and Genre')

# Show the plot
plt.show()
```

#India has the highest number of content available to it

#Analysis of actors/directors of different types of shows/movies.

```python
# Split the "director" column into multiple rows based on commas
df['directors_split'] = df['director'].apply(lambda x:str(x).split(",")).tolist()


# Display the updated DataFrame
df.head()
```

```python
# Count the occurrences of movies and TV shows directed by Christopher Nolan
nolan_movies_count = df[(df['directors_split'].apply(lambda x: 'Christopher Nolan' in x)) & (df['type'] == 'Movie')].shape[0]
nolan_tv_shows_count = df[(df['directors_split'].apply(lambda x: 'Christopher Nolan' in x)) & (df['type'] == 'TV Show')].shape[0]

# Compare the counts
if nolan_movies_count > nolan_tv_shows_count:
    print("Christopher Nolan has made more movies.")
elif nolan_movies_count < nolan_tv_shows_count:
    print("Christopher Nolan has made more TV shows.")
else:
    print("Christopher Nolan has made an equal number of movies and TV shows.")
```

```python
# Create a bar plot to compare the counts
labels = ['Movies', 'TV Shows']
counts = [nolan_movies_count, nolan_tv_shows_count]

plt.bar(labels, counts,color ='red')
plt.xlabel('Content Type')
plt.ylabel('Count')
plt.title('Number of Movies vs TV Shows Directed by Christopher Nolan')

plt.show()
```

#Hence we can infer that christopher nolan has made more movies .

```python
# Count the occurrences of each director
director_counts = df['director'].value_counts().head(10)

# Create a bar plot of the top 10 directors
plt.bar(director_counts.index, director_counts.values,color ='red')
plt.xlabel('Director')
plt.ylabel('Number of Movies')
plt.title('Top 10 Directors with the Most Movies')

# Rotate x-axis labels for better visibility
plt.xticks(rotation=45, ha='right')

# Display the plot
plt.show()
```

#Actor analysis


```python
# Convert "cast" and "type" columns to string if they contain float values
if df['cast'].dtype == float:
    df['cast'] = df['cast'].astype(str)

if df['type'].dtype == float:
    df['type'] = df['type'].astype(str)

# Split the "cast" column into individual actors
df['actors'] = df['cast'].str.split(', ')

# Explode the DataFrame to create separate rows for each actor
df_exploded = df.explode('actors')

# Count the occurrences of each actor for movies
movie_actor_counts = df_exploded[df_exploded['type'] == 'Movie']['actors'].value_counts().head(10)

# Count the occurrences of each actor for TV shows
tvshow_actor_counts = df_exploded[df_exploded['type'] == 'TV Show']['actors'].value_counts().head(10)

# Create a bar plot for movies
plt.bar(movie_actor_counts.index, movie_actor_counts.values, label='Movies')

# Create a bar plot for TV shows
plt.bar(tvshow_actor_counts.index, tvshow_actor_counts.values, label='TV Shows')

plt.xlabel('Actor')
plt.ylabel('Number of Occurrences')
plt.title('Top 10 Actors in Movies and TV Shows')
plt.legend()

# Rotate x-axis labels for better visibility
plt.xticks(rotation=45, ha='right')

# Display the plot
plt.show()
```
